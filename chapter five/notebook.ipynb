```python
# in this case, every levl from DEBUG to CRITICAL will be logged
logging.basicConfig(level=logging.DEBUG)
```

```python
# location - if you want the logs on a separate file:
logging.basicConfig(file_name="chapter_5_logs.log", level=logging.DEBUG)
```
