## pandas dataframes

```python
# creating a series
usa_data = pd.Series(
    [13.33, 14.02, 14.02, 14.25],
    index = ["2000", "2001", "2002", "2003"]
)
```

```python
usa_data
```

```python
# constructing a pandas dataframe
kenya_data = pd.Series(
    [9.02, 9.01, 8.84, 8.84],
    index = ["2000", "2001", "2002", "2003"]
)

df = pd.DataFrame(
    {
        "USA": usa_data,
        "Kenya": kenya_data
    }
)
```

```python
df
```

```python
%%timeit
df["Kenya_fraction"] = df["Kenya"] / 100
```

```python
%%timeit
df["Kenya_fraction"] = df["Kenya"].apply(lambda x: x / 100)
```
